```python
import requests
from github import Github
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt
import seaborn as sb

# Pull Request class for required fields
class PullRequests:
    # initializing the variables with None
    def __init__(self):
        self.user = None
        self.repository = None
        self.title = None
        self.created_time = None
        self.body = None
        self.state = None
        self.commits = None
        self.additions = None
        self.deletions = None
        self.files_changed = None

# Repositories class for required fields
class Repositories:
    # initializing the variables with None
    def __init__(self):
        self.user = None
        self.full_name = None
        self.description = None
        self.creation_time = None
        self.last_push = None
        self.No_of_users = None

# User class for required fields
class Users:
    # initializing the variables with None
    def __init__(self):
        self.name = None
        self.followers = None
        self.following = None
        self.contributions = None
        self.pullrequests = None

# Class to print the output
class PrintInfo:

    # Method to check empty csv file
    def is_empty_csv(self, name):
        with open(name) as csvfile:
            reader = csv.reader(csvfile)
            for i, j in enumerate(reader):
                if i and j and str(j).strip():  # found the second row
                    return False
        return True

    # Print Pull request data from csv file
    def print_pullRequest(self, user, repo, title):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        print('-'*100)
        df = PDF[(PDF.user == user) & (PDF.repository == repo) & (PDF.title == title)]
        if len(df) == 0:
            print('No Data')
            return
        for i in range(len(df)):
            print('User\t\t: ',df.iloc[i, 0])
            print('Repository\t: ', df.iloc[i, 1])
            print("Title\t\t: ", df.iloc[i, 2])
            print("Created Date\t: ", df.iloc[i, 3])
            print("Body\t\t: ", df.iloc[i, 4])
            print("State\t\t: ", df.iloc[i, 5])
            print("Commits\t\t: ", df.iloc[i, 6])
            print("Additions\t: ", df.iloc[i, 7])
            print("Deletions\t: ", df.iloc[i, 8])
            print("Changes Files\t: ", df.iloc[i, 9])
            print('-'*100)
    
    # Print Repository Data from csv file
    def print_repositry(self, user, repo=None):
        if self.is_empty_csv('Repositories.csv'):
            print('Repositories File is empty')
            return
        print('-'*100)
        pdf = pd.read_csv('Repositories.csv')
        df = pdf[(pdf.user == user) & (pdf.full_name == repo)]
        if len(df) == 0:
            print('No Data')
            return
        for i in range(len(df)):
            print('User\t\t: ',df.iloc[i, 0])
            print('Name\t\t: ',df.iloc[i, 1])
            print('Description\t: ',df.iloc[i, 2])
            print('Creation Time\t: ',df.iloc[i, 3])
            print('Last Push\t: ',df.iloc[i, 4])
            print('No_of_users\t: ',df.iloc[i, 5])
            print('-'*100)

    # Print Users Data from csv file
    def print_user(self, user):
        if self.is_empty_csv('Users.csv'):
            print('Users File is empty')
            return
        print('-'*100)
        pdf = pd.read_csv('Users.csv')
        df = pdf[(pdf.name == user)]
        if len(df) == 0:
            print('No Data')
            return
        for i in range(len(df)):
            print('Name\t\t: ', df.iloc[i, 0])
            print('Followers\t: ', df.iloc[i, 1])
            print('Following\t: ', df.iloc[i, 2])
            print('Contributions\t: ', df.iloc[i, 3])
            print('Pull requests\t:', df.iloc[i,4])
            print('-'*100)

    # Method to display the users we fetched from git
    def get_user_names(self):
        if self.is_empty_csv('Users.csv'):
            print('Users File is empty')
            return
        print('-'*100)
        pdf = pd.read_csv('Users.csv', usecols=['name'])
        if len(pdf) == 0:
            print('No Data')
            return
        print("Available User Names are : ")
        print(pdf)
        print('-'*100)

    # Method to display the list of repositories we stored
    def get_repos(self):
        if self.is_empty_csv('Repositories.csv'):
            print('Repositories File is empty')
            return
        print('-'*100)
        pdf = pd.read_csv('Repositories.csv', usecols =['user','full_name'])
        print("Available Repository Names are : ")
        if len(pdf) == 0:
            print('No Data')
            return
        print(pdf)
        print('-'*100)

    # Method to display the list of PRs we stored
    def get_PR_details(self):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        print('-'*100)
        df = pd.read_csv('pullRequests.csv', usecols=['user', 'repository', 'title'])
        if len(df) == 0:
            print('No Data')
            return
        print('Available PRs are : ')
        print('User\t\t\t\t\tRepository\t\t\t\t\tTitle')
        for i in range(len(df)):
            print(df.iloc[i, 0], df.iloc[i, 1], df.iloc[i, 2], sep = '\t\t\t')
        print('-'*100)

    # Method to display the count of open Pull Requests in a repository
    def get_open_PR_count(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        pdf = pd.read_csv('pullRequests.csv')
        df = pdf[(pdf.user == user) & (pdf.repository == repo)]
        count = 0
        for i in range(len(df)):
            if df.iloc[i,5] == 'open':
                count+=1
        print("No of open PRs for repository",repo,':',count)
        return count

    def get_total_PR_count(self, user):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        pdf = pd.read_csv('pullRequests.csv')
        df = pdf[(pdf.user == user)]
        count = 0
        for i in range(len(df)):
                count+=1
        return count

    # Method to display the count of closed Pull Requests in a repository
    def get_closed_PR_count(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        pdf = pd.read_csv('pullRequests.csv')
        df = pdf[(pdf.user == user) & (pdf.repository == repo)]
        count = 0
        for i in range(len(df)):
            if df.iloc[i,5] == 'closed':
                count+=1
        global closed_PR_count 
        closed_PR_count = count
        
        print("No of Closed PRs for repository",repo,':',count)
        return count

    # Method to display the count of stored users
    def get_user_count(self, user, repo):
        if self.is_empty_csv('Repositories.csv'):
            print('Repositories File is empty')
            return
        pdf = pd.read_csv('Repositories.csv')
        df = pdf[(pdf.user == user) & (pdf.full_name == repo)]
        global user_count
        user_count = df.iloc[0, 5]
        #global user_count
        print("Number of Users are : ", df.iloc[0, 5])

    # Method to display the time stamp of oldest PR of a reository
    def oldest_PR(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        df = PDF[(PDF.user == user) & (PDF.repository == repo)]
        if len(df) == 0:
            print('No Data')
            return
        print("Oldest PR time is : ", max(df['created_time']))

    # Method that compares closed vs open pull requests in terms of commits and draw boxplot
    def boxplot_commits(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        df = PDF[(PDF.user == user) & (PDF.repository == repo)]
        df.fillna(0)
        x=df[(df.state =='open')]
        x.rename(columns = {'commits':'OpenCommits'}, inplace = True)
        x=x.OpenCommits
        y=df[(df.state =='closed')]
        y.rename(columns = {'commits':'ClosedCommits'}, inplace = True)
        y=y.ClosedCommits
        z=pd.concat([x,y],axis=1)
        z=z.fillna(0)
        if len(z) == 0:
            print('No Data')
            return
        sb.boxplot(data=[z["OpenCommits"], z["ClosedCommits"]])
        plt.title('compares closed vs open pull requests in terms of commits')
        plt.show()

    # Method that compares closed vs open pull requests in terms of additions and draw boxplot
    def boxplot_additions(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        df = PDF[(PDF.user == user) & (PDF.repository == repo)]
        #df.fillna(0)
        x=df[(df.state =='open')]
        x.rename(columns = {'additions':'Openadditions'}, inplace = True)
        x=x.Openadditions
        y=df[(df.state =='closed')]
        y.rename(columns = {'additions':'Closedadditions'}, inplace = True)
        y=y.Closedadditions
        z=pd.concat([x,y],axis=1)
        z=z.fillna(0)
        if len(z) == 0:
            print('No Data')
            return
        sb.boxplot(data=[z["Openadditions"], z["Closedadditions"]])
        plt.title('compares closed vs open pull requests in terms of additions')
        plt.show()

    # Method that compares closed vs open pull requests in terms of deletions and draw boxplot
    def boxplot_deletions(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        df = PDF[(PDF.user == user) & (PDF.repository == repo)]
        #df.fillna(0)
        x=df[(df.state =='open')]
        x.rename(columns = {'deletions':'Opendeletions'}, inplace = True)
        x=x.Opendeletions
        y=df[(df.state =='closed')]
        y.rename(columns = {'deletions':'Closeddeletions'}, inplace = True)
        y=y.Closeddeletions
        z=pd.concat([x,y],axis=1)
        z=z.fillna(0)
        if len(z) == 0:
            print('No Data')
            return
        sb.boxplot(data=[z["Opendeletions"], z["Closeddeletions"]])
        plt.title('compares closed vs open pull requests in terms of deletions')
        plt.show()

    # Method that compares relationship between additions and deletions and draw boxplot
    def scatterplot(self, user, repo):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        PDF = pd.read_csv('pullRequests.csv')
        df = PDF[(PDF.user == user) & (PDF.repository == repo)]
        #df.fillna(0)
        df = df[['additions','deletions']]
        if len(df) == 0:
            print('No Data')
            return
        plt.title('compares relationship between additions and deletions')
        plt.scatter(df["additions"],df["deletions"])
        plt.show()

    # Method showing number of pull requests pe day in graph format
    def plot_PR_day_graph(self):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        df = pd.read_csv('pullRequests.csv')
        df['created_time'] = pd.to_datetime(df['created_time']).dt.date
        #df.fillna(0)
        d={}
        for i in range(len(df)):
            if str(df.iloc[i,3]) in d:
                d[str(df.iloc[i,3])]+=1
            else:
                d[str(df.iloc[i,3])]=1
        if len(d) == 0:
            print('No Data')
            return
        plt.title('number of pull requests per day')
        plt.xticks(rotation='vertical')
        plt.plot(list(d.keys()), list(d.values()))
        plt.xticks(rotation='vertical')
        plt.show()

    # Method showing number of pull open requests pe day in graph format
    def plot_OPR_day_graph(self):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        df = pd.read_csv('pullRequests.csv')
        df=df[(df.state =='open')]
        df['created_time'] = pd.to_datetime(df['created_time']).dt.date
        #df.fillna(0)
        d={}
        for i in range(len(df)):
            if str(df.iloc[i,3]) in d:
                d[str(df.iloc[i,3])]+=1
            else:
                d[str(df.iloc[i,3])]=1
        if len(d) == 0:
            print('No Data')
            return
        plt.title('number of pull open requests per day')
        plt.plot(list(d.keys()), list(d.values()))
        plt.xticks(rotation='vertical')
        plt.show()

    # Method showing number of pull closed requests pe day in graph format
    def plot_CPR_day_graph(self):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        df = pd.read_csv('pullRequests.csv')
        df=df[(df.state =='closed')]
        df['created_time'] = pd.to_datetime(df['created_time']).dt.date
        #df.fillna(0)
        d={}
        for i in range(len(df)):
            if str(df.iloc[i,3]) in d:
                d[str(df.iloc[i,3])]+=1
            else:
                d[str(df.iloc[i,3])]=1
        if len(d) == 0:
            print('No Data')
            return
        plt.title('number of pull closed requests per day')
        plt.plot(list(d.keys()), list(d.values()))
        plt.xticks(rotation='vertical')
        plt.show()

    # Method comparing users per repository in bar graph
    def plot_bar(self):
        if self.is_empty_csv('Repositories.csv'):
            print('Repositories File is empty')
            return
        df = pd.read_csv('Repositories.csv')
        if len(df) == 0:
            print('No Data')
            return
        plt.bar(df["full_name"],df["No_of_users"])
        plt.title('comparing users per repository')
        plt.xlabel('name')
        plt.ylabel('No of users')
        plt.xticks(rotation='vertical')
        plt.show()

    # Method to print Correlation between numeric values of pull requests
    def Correlate_PR(self):
        if self.is_empty_csv('pullRequests.csv'):
            print('pullRequests File is empty')
            return
        df = pd.read_csv('pullRequests.csv')
        df = df[['commits','additions','deletions','files_changed']]
        if len(df) == 0:
            print('No Data')
            return
        df = df.corr()
        print(df)

    def Correlate_user(self):
        if self.is_empty_csv('Users.csv'):
            print('Users File is empty')
            return
        df = pd.read_csv('Users.csv')
        df = df[['followers','following','contributions', 'pullrequests']]
        if len(df) == 0:
            print('No Data')
            return
        print(df.corr())

class Process:
    # Method to insert the data into class object then save it to PulRequests.csv file
    def Set_pullRequest(self, pull, repname, user):
        pullreq = PullRequests()
        pullreq.user = user
        pullreq.repository = repname
        pullreq.title = pull.title
        pullreq.created_time =  pull.created_at
        pullreq.body =  pull.body
        pullreq.state = pull.state
        pullreq.commits = pull.commits
        pullreq.additions = pull.additions
        pullreq.deletions = pull.deletions
        pullreq.files_changed = pull.changed_files

        #fetching the data from csv and appending the new data
        PDF = pd.read_csv('pullRequests.csv')
        dt = pullreq.__dict__
        hv = pd.DataFrame(dt,index=['h'])
        newDf = pd.concat([PDF,hv],axis = 0)

        #Updating the csv with new data
        newDf.to_csv('pullRequests.csv',index = False)
        pinfo = PrintInfo()
        pinfo.print_pullRequest(user, repname, pullreq.title)

    # Method to insert the data into class object then save it to Repositories.csv file
    def set_repo(self, repo, user):
        rep = Repositories()
        rep.user = user
        rep.full_name = repo.full_name
        rep.creation_time = repo.created_at
        rep.last_push = repo.pushed_at
        rep.description = repo.description

        # fetching open pull requests
        Pulls = repo.get_pulls()
        for pull in Pulls:
            self.Set_pullRequest(pull, repo.full_name, user)

        # fetching closed pull requests
        Pulls = repo.get_pulls(state='closed')
        for pull in Pulls:
            self.Set_pullRequest(pull, repo.full_name, user)

        # fetching contributors
        count = 0
        for i in repo.get_contributors():
            count+=1
        rep.No_of_users = count

        #fetching the data from csv and appending the new data
        PDF = pd.read_csv('Repositories.csv')
        dt = rep.__dict__
        hv = pd.DataFrame(dt,index=['h'])
        newDf = pd.concat([PDF,hv],axis = 0)

        #Updating the csv with new data
        newDf.to_csv('Repositories.csv',index = False)
        pinfo = PrintInfo()
        pinfo.print_repositry(user,repo.full_name)
    
    # Method to insert the data into class object then save it to Users.csv file
    def set_user(self, guser):
        user = Users()
        user.name = guser.name
        user.followers = guser.followers
        user.following = guser.following
        user.contributions = guser.contributions

        pinfo = PrintInfo()
        user.pullrequests = pinfo.get_total_PR_count(guser.name)

        #fetching the data from csv and appending the new data
        PDF = pd.read_csv('Users.csv')
        dt = user.__dict__
        hv = pd.DataFrame(dt,index=['h'])
        newDf = pd.concat([PDF,hv],axis = 0)

        #Updating the csv with new data
        newDf.to_csv('Users.csv',index = False)
        pinfo.print_user(guser.name)

    # Fetching the data from Github
    def fetch_git_info(self):
        username = input("Enter User Name : ")
        pinfo = PrintInfo()
        user = g.get_user(username)
        reponame = input("Enter Repository Name : ")
        repo = user.get_repo(reponame)
        p.set_repo(repo, user.name)

        self.set_user(user)

usr = Users()
rep = Repositories()
pr = PullRequests()

dt1 = usr.__dict__
dt2 = rep.__dict__
dt3 = pr.__dict__

with open('pullRequests.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(dt3.keys())
with open('Repositories.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(dt2.keys())
with open('Users.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(dt1.keys())

g = Github("") # using the github token
p= Process()
q = PrintInfo()
while True: # entering into while loop
    print('1.Fetch User info\n2.Print User Data\n3.Print Repository Data\n4.Print Pull Request Data\n5.Repository Summary')
    print('6.Bolplot Graphs and Scatter graph\n7.line graphs\n8.Correlation\n9.Exit')
    opt = int(input("Enter Your choice : "))
    if opt == 1:
        p.fetch_git_info()
    elif opt == 2:
        q.get_user_names()
        global name
        name = input("Enter User Name : ")
        
        q.print_user(name)
    elif opt == 3:
        q.get_repos()
        name = input("Enter User Name : ")
        global repo
        repo = input("Enter Repository : ")
       
        q.print_repositry(name, repo)
    elif opt == 4:
        q.get_PR_details()
        name = input("Enter User Name : ")
        repo = input("Enter Repository : ")
        title = input("Enter PR Title :")
        q.print_pullRequest(name, repo, title)
    elif opt == 5:
        q.get_repos()
        name = input("Enter User Name : ")
        repo = input("Enter Repository : ")
        print('-'*100)
        q.get_closed_PR_count(name, repo)
        q.get_open_PR_count(name, repo)
        q.get_user_count(name, repo)
        q.oldest_PR(name, repo)
    elif opt == 6:
        q.get_repos()
        name = input("Enter User Name : ")
        repo = input("Enter Repository : ")
        q.boxplot_commits(name,repo)
        q.boxplot_additions(name,repo)
        q.boxplot_deletions(name,repo)
        q.scatterplot(name,repo)
    elif opt == 7:
        q.plot_PR_day_graph()
        q.plot_OPR_day_graph()
        q.plot_CPR_day_graph()
        q.plot_bar()
    elif opt == 8:
        print('User Correlation :')
        q.Correlate_user()
        print('-'*100)
        print('\nPR correlation')
        q.Correlate_PR()
    else:
        break
    print('='*100)

```
