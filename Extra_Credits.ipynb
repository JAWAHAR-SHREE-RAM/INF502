```python
import pandas as pd # Importing  for data analysis
import requests # Importing  to handle requests
from bs4 import BeautifulSoup # Importing  to parse HTML documents
```

```python

wikiurl="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population_density" # Decalring the url 
response=requests.get(wikiurl) # geting into  url
print(response.status_code) # Displaying the status code
```

```python
soup = BeautifulSoup(response.text, 'html.parser') #Using the  HTML string to be parsed.
table=soup.find('table',{'class':"wikitable"})
table # Displaying the the table in html format 
```

```python
df=pd.read_html(str(table))
df=pd.DataFrame(df[0]) # Converting in to dataframe 
df.head() # Dispalying the dataset df
```
